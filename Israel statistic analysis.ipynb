```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\user\anaconda3\pkgs\tensorflow-base-2.6.0-gpu_py39hb3da07e_0\Lib\site-packages\tensorflow\python\keras\datasets\Israel-Palestine.csv")
```

```python
data
```

```python
data = data[data['Country']=='Israel']
```

```python
data
```

```python
data.columns
```

```python
for i in ['GDP (in USD)','Agricultural Output (in USD)','IT Output (in USD)']:
    data[i] = data[i].apply(lambda x:float(x.split(' ')[0])*1000000000)
```

```python
data
```

```python
data = data.set_index('Year')
```

```python
data
```

```python
for i in ['Population','Active Military Personnel','Reserve Military Personnel','Number of Tanks','Number of Armoured Vehicles']:
    data[i] = data[i].str.replace(',','')
    data[i] = data[i].astype('float')
```

```python

for i in data.drop('Country', axis =1 ).columns:
    data[i].plot(kind = 'line')
    plt.title(i)
    plt.grid()
    plt.tight_layout()
    plt.show()
```
